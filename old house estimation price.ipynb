```python
from tkinter import *
import mysql.connector
from tkinter import ttk

mydb=mysql.connector.connect(
    host="localhost",
    user="root",
    password="12345",
    database="tushar",
    autocommit=True
)
my_cursor=mydb.cursor()
top=Tk()
top.geometry("500x400")
top.title('PRICE ESTIMATION OF USED HOUSES')
top.configure(bg="red")

frame=Frame(bg="yellow")
var=IntVar()
st=StringVar()
sto=StringVar()
bed=StringVar()
bath=StringVar()
sq=StringVar()

dictu={'Bangalore':20146,'Chennai':3331,'Delhi':20000,'Ghaziabad':12674,'Gurgaon':25000,
                    'Hyderabad':10000,'Kolkata':17000,'Mumbai':70193,'Noida':27075,'Pune':12000,'Jalandhar':4500,
                    'Lucknow':26492,'Goa':23333,'Jaipur':17000,'Kochi':8000,'Patna':6946,'Agra':31515,'Meerut':19667,
                    'Ranchi':7917,'Ludhiyana':10667}

dictr={'Bangalore':6146,'Chennai':5331,'Delhi':15000,'Ghaziabad':7674,'Gurgaon':15000,
                    'Hyderabad':7000,'Kolkata':6000,'Mumbai':50193,'Noida':17075,'Pune':9000,'Jalandhar':2000,
                    'Lucknow':12492,'Goa':10333,'Jaipur':10500,'Kochi':4000,'Patna':4000,'Agra':8515,'Meerut':11667,
                    'Ranchi':4917,'Ludhiyana':4667}

def calculate():
    
    e1=ent1.get()
    e2=ent2.get()
    e3=ent3.get()
    e4=ent4.get()
    e5=ent5.get()
    pr=0
    brick=0
    bat=0
    if(var.get()==1):
        
        for ci in dictr:
            if(e1==ci):
                if(int(e2)==1):
                    if(int(e3)==1):
                        brick=20000
                    elif(int(e3)==2):
                        brick=40000
                    elif(int(e3)==3):
                        brick=50000
                    elif(int(e3)==4):
                        brick=60000
                if(int(e4)==1):
                    bat=5000
                if(int(e4)==2):
                    bat=10000
                        
                pr=(int(e5)*dictr[e1]+brick+bat)
                

    else:
        for ci in dictu:
            if(e1==ci):
                if(int(e2)==1):
                    if(int(e3)==1):
                        brick=30000
                    elif(int(e3)==2):
                        brick=50000
                    elif(int(e3)==3):
                        brick=60000
                    elif(int(e3)==4):
                        brick=70000
                if(int(e4)==1):
                    bat=7000
                if(int(e4)==2):
                    bat=14000
                        
                pr=(int(e5)*dictu[e1]+brick+bat)
                
            
            
    cmd="INSERT INTO HOUSES(STATE,STOREROOMS,BEDROOMS,BATHROOMS,SQFT,PRICE) VALUES(%s,%s,%s,%s,%s,%s)"
    val=(e1,e2,e3,e4,e5,str(pr))
    #l="SELECT * FROM OLDHOUSES"
    my_cursor.execute(cmd,val)
    
    ff=Frame(bg="purple")
    lab=Label(ff,text="The estimated Price of this catagory is "+str(pr),font=('Ariel',13,'bold'),bg="purple")
    lab.grid(row=25,column=3)
    ff.pack(side=TOP)
    
def widget():
    global ent1,ent2,ent3,ent4,ent5
    lab1=Label(frame,text="City",font=('Helvetica',18,'bold'),bg="yellow")
    lab1.grid(row=3,column=1)
    ent1=ttk.Combobox(frame,width=27,textvariable=st)
    ent1['values']=('Bangalore','Chennai','Delhi','Ghaziabad','Gurgaon',
                    'Hyderabad','Kolkata','Mumbai','Noida','Pune','Jalandhar',
                    'Lucknow','Goa','Jaipur','Kochi','Patna','Agra','Meerut',
                    'Ranchi','Ludhiyana')
    ent1.grid(row=3,column=12)
    
    lab2=Label(frame,text="   Store rooms   ",font=('Helvetica',18,'bold'),bg="yellow")
    lab2.grid(row=5,column=1)
    ent2=ttk.Combobox(frame,width=27,textvariable=sto)
    ent2['values']=('1')
    ent2.grid(row=5,column=12)
    
    lab3=Label(frame,text="    Bedrooms    ",font=('Helvetica',18,'bold'),bg="yellow")
    lab3.grid(row=7,column=1)
    ent3=ttk.Combobox(frame,width=27,textvariable=bed)
    ent3['values']=('1','2','3','4')
    ent3.grid(row=7,column=12)
    
    lab4=Label(frame,text="Bathrooms",font=('Helvetica',18,'bold'),bg="yellow")
    lab4.grid(row=9,column=1)
    ent4=ttk.Combobox(frame,width=27,textvariable=bath)
    ent4['values']=('1','2')
    ent4.grid(row=9,column=12)
    
    lab5=Label(frame,text="      SQFt      ",font=('Helvetica',18,'bold'),bg="yellow")
    lab5.grid(row=11,column=1)
    ent5=ttk.Combobox(frame,width=27,textvariable=sq)
    ent5['values']=('300','400','600','635','700','1000')
    ent5.grid(row=11,column=12)
    
    
    
    
    bt=Button(frame,text="Submit",command=calculate,bg="aqua")
    bt.grid(row=17,column=12)
    
    
    
r=Frame(bg="red",width=500)
lab=Label(r,text="         WELCOME",font=('Helvetica',12,'bold'),bg="red")
lab.grid(row=0,column=4)
rad=Radiobutton(r,text="Rural",variable=var,value=1,command=widget,bg="red")

rad.grid(row=1,column=3)
rad1=Radiobutton(r,text="Urban",variable=var,value=2,command=widget,bg="red")
rad1.grid(row=1,column=6)
r.pack()
frame.config()
frame.pack(side=TOP)
top.mainloop()
```
